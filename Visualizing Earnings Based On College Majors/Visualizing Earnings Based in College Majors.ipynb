Exploring relations between Sample_Size and Median

```python
recent_grads.plot(x='Sample_size',y='Median',title='Sample_size vs Median',kind='scatter')
```

There seems to be a negative correlation between Sample size and Median. But there's a cluster that's of interest.

```python
recent_grads.plot(x='Sample_size',y='Unemployment_rate',title='Sample_size vs Unemployment_rate',kind='scatter')
```

Again, there seems to be a negative correlation between Sample size and Unemployment rate. But there's a cluster that's of interest which needs to be looked into.

```python
recent_grads.plot(x='Full_time',y='Median',title='Full_time vs Median',kind='scatter')
```

Yet again, there seems to be a negative correlation between Sample size and Unemployment rate. But there's a cluster that's of interest which needs to be looked into.

```python
recent_grads.plot(x='ShareWomen',y='Unemployment_rate',title='ShareWomen vs Unemployment_rate',kind='scatter')
```

There doesn't seem to be any correlation between Sharewomen and Unemployment_rate.

```python
recent_grads.plot(x='Men',y='Median',title='Men vs Median',kind='scatter')
```

Similar to the first four comparisions, there seems to be a negative correlation between Men and Median.

```python
recent_grads.plot(x='Women',y='Median',title='Women vs Median',kind='scatter')
```

Similar to the last comparision, there seems to be a negative correlation between Men and Median.

```python
recent_grads.plot(x='Total',y='Median',title='Total vs Median',kind='scatter')
```

Surprisingly, the more popular a major is (the more a major is pursued) the lesser its median.

```python
maj_fem = recent_grads[recent_grads['Women']>recent_grads['Men']]
maj_mal = recent_grads[recent_grads['Men']>recent_grads['Women']]
fem_mean = maj_fem['Median'].mean()
men_mean = maj_mal['Median'].mean()
print(fem_mean,men_mean)
```

Students that majored in subjects that were majority male seem to make more money than students that majored in subjects that were majority female.

```python
recent_grads.plot(x='Full_time',y='Median',kind='scatter')
```
