# KMeans Clustering

```python
import pandas as pd
import numpy as np
from bokeh.io import output_notebook, show, curdoc, push_notebook
from bokeh.models import ColumnDataSource,HoverTool,ColorBar
from bokeh.plotting import figure, show, output_notebook
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from bokeh.layouts import layout
from bokeh.layouts import row
from ipywidgets import interact
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual, IntSlider

output_notebook()

data = pd.read_csv('Wholesale customers data.csv')
```

```python
data = data.drop('Channel',axis=1)
data = data.drop('Region',axis=1)
data.head(10)
```

### Dataset loaded into dataframe

### Creating the KMeans model, bokeh figure and plotting circle glyphs

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[['Milk','Grocery']])
data['cluster'] = kmeans.labels_

colormap = {0: 'red', 1: 'blue', 2: 'green', 3: 'yellow', 4: 'orange', 5: 'black', 6: 'navy', 7: 'pink'}
colors = [colormap[x] for x in data['cluster']]

p = figure(title = "Wholesale Customers Data", tools="hover,lasso_select,pan,wheel_zoom,box_zoom,reset,save")
p.xaxis.axis_label = 'Milk'
p.yaxis.axis_label = 'Grocery' 

x = data['Milk'].tolist()
y = data['Grocery'].tolist()

r = p.circle(x, y, color=colors, fill_alpha=0.2, size=10)
```

### Defining method for updating plot based on user interaction

```python
def update(X_Axis,Y_Axis,Number_of_Clusters,Max_Iterations):
    kmeans_new = KMeans(n_clusters=Number_of_Clusters,max_iter=Max_Iterations)
    kmeans_new.fit(data[[X_Axis,Y_Axis]])
    data['cluster_updated'] = kmeans_new.labels_

    colormap_new = {0: 'red', 1: 'blue', 2: 'green', 3: 'yellow', 4: 'orange', 5: 'black', 6: 'navy', 7: 'pink'}
    colors_new = [colormap_new[x] for x in data['cluster_updated']]
    
    x_new = data[X_Axis].tolist()
    y_new = data[Y_Axis].tolist()
        
    r.data_source.data['x'] = data[X_Axis]
    r.data_source.data['y'] = data[Y_Axis]
    p.xaxis.axis_label = X_Axis
    p.yaxis.axis_label = Y_Axis
    r.data_source.data['fill_color'] = colors_new
    r.data_source.data['line_color'] = colors_new
    
    push_notebook()
```

```python
show(p, notebook_handle=True)
```

### Please use interactive tools below to modify axes, number of clusters and maximum iterations

```python
interact(update, X_Axis = list(data.columns)[:-1], Y_Axis = list(data.columns)[:-1], Number_of_Clusters = IntSlider(min=1,max=8,step=1,value=4), Max_Iterations = IntSlider(min=1,max=1000,step=1,value=300))
```

### The KMeans clustering algorithm calculates centroids, one for each cluster, and group data points based on their distances from these centroids.

### When zoomed in, it can be noticed that as the number of clusters increase, the number of iterations have to be increased to reach an optimum value.
